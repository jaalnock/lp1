```python
# Hierarchical Clustering
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the dataset
df = pd.read_csv('./Datasets/50_Startups.csv')

# Extract the feature for clustering
X = df[['PROFIT']]

# Standardize the data
X_scaled = StandardScaler().fit_transform(X)

# Apply Agglomerative Clustering
cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
df['cluster'] = cluster.fit_predict(X_scaled)

# Visualize the clusters
sns.scatterplot(x='PROFIT', y='RND', hue='cluster', data=df)
plt.title('Hierarchical Clustering - Profit vs. R&D Spending')
plt.xlabel('Profit')
plt.ylabel('R&D Spending')
plt.show()

# Create and visualize the dendrogram
dendrogram(linkage(X_scaled, method='ward'))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Point Index')
plt.ylabel('Distance')
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('./Datasets/50_Startups.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
X = data[['PROFIT']]
y = data['RND']
```

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

dendrogram(linkage(X_scaled, method='ward'))
plt.title('Dendrogram for finding optimal no. of Clusters')
plt.xlabel('Profit')
plt.ylabel('Eucledian Distance')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hm = AgglomerativeClustering(n_clusters=3)
labels = hm.fit_predict(X_scaled)
labels
```

```python

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.title("Unclustered Data")
plt.xlabel('Profit')
plt.ylabel('RND')
plt.scatter(X, y)

plt.subplot(1,2,2)
plt.title("Clustered Data")
plt.xlabel('Profit')
plt.ylabel('RND')
plt.scatter(X, y, c=labels)
```
